```python
df['income_category'] = pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])
df.head()
```

```python
df['income_category'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title('Income Category Distribution')
plt.xlabel('Income Category')
plt.ylabel('Number of Instances')
```

```python
#STRATIFIED_SHUFFLE_SPLIT
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_category']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_test_set['income_category'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title('Income Category Distribution')
plt.xlabel('Income Category')
plt.ylabel('Number of Instances')
```

```python
strat_train_set['income_category'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title('Income Category Distribution')
plt.xlabel('Income Category')
plt.ylabel('Number of Instances')
```

# Data Cleaning and Preprocessing

```python
df = strat_train_set.copy()
df
```

```python
housing = strat_train_set.drop('median_house_value', axis=1)
housing_labels = strat_train_set['median_house_value'].copy()
housing_labels
```

## Handling missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = 'median')

housing_num = housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)
```

```python
X = imputer.transform(housing_num)
```

```python
X
```

```python
housing = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index)
housing
```

# Handling Categorical data

```python
housing['ocean_proximity'] = df['ocean_proximity']
housing
```

```python
housing = housing[['ocean_proximity']]
housing
set(housing['ocean_proximity'])
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder()
housing_cat = one_hot_encoder.fit_transform(housing)
housing_cat.toarray()
```

```python
housing_cat = pd.DataFrame(housing_cat.toarray(), columns=['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN'], index=housing.index)
```

```python
housing_cat
```

```python
# Ensure no duplicate columns before concat
df = df.loc[:, ~df.columns.isin(housing_cat.columns)]

# Concatenate cleanly
df = pd.concat([df, housing_cat], axis=1)
df = df.drop('ocean_proximity', axis=1)
df
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_sclaed = scaler.fit_transform(df)
df_sclaed = pd.DataFrame(df_sclaed, columns=df.columns, index=df.index)
df_sclaed
```
